```python
#Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#Loading the Dataset
epl = pd.read_csv('EPL_20_21.csv')
```

```python
#Checking the first 5 rows of the dataset
epl.head()
```

```python
#to check the total number of columns and rows in the dataset
epl.info()
```

```python
epl.describe()
```

```python
#to find total number of null values
epl.isna().sum()
```

```python
#Create new columns
epl['Goals_per_match'] = epl['Goals']/epl['Matches'].astype(float)
epl['Mins_per_match'] = epl['Mins']/epl['Matches'].astype(float)
```

```python
epl.head()
```

```python
#Total number of goals scored in the season 
total_goals = epl['Goals'].sum()
```

```python
#Total number of penalties scored in the season
total_pens = epl['Penalty_Goals'].sum()
print(total_pens)
```

```python
#Total penalty attempts 
total_pens_att = epl['Penalty_Attempted'].sum()
print(total_pens_att)
```

```python
#Penalties missed
total_pens_missed = total_pens_att - total_pens
print(total_pens_missed)
```

```python
#Visualise pens scored vs missed using pie chart
plt.figure(figsize =(12,4))
data = [total_pens_missed,total_pens]
labels = ['Penalties Missed','Penalties Scored']
plt.pie(data,labels = labels,autopct = '%.0f%%')
```

```python
#No of unique positions of players
epl['Position'].nunique()
```

```python
#No of forwards players 
epl[epl['Position'] == 'FW'].count()
```

```python
#No of nationalities in the league
epl['Nationality'].nunique()
```

```python
#Player distribution by countries
nationality = epl['Nationality'].value_counts().head(10)
nationality.plot(kind = 'bar',figsize = (12,4))
nationality.xlabel = 'No. of players'
```

```python
#Top 5 clubs with most players
epl['Club'].value_counts().head(5).plot(kind = 'bar', color = 'Green')
```

```python
#Clubs with least number of players
epl['Club'].value_counts().tail(5).plot(kind = 'bar', color = 'Red')
```

```python
#Player based on Age Group
Under20 = epl[epl['Age'] <= 20]
Age20_25 = epl[(epl['Age'] > 20) & (epl['Age'] <= 25)]
Age25_30 = epl[(epl['Age'] > 25) & (epl['Age'] <= 30)]
Above30 = epl[epl['Age'] > 30]
```

```python
x = np.array([Under20['Name'].count(), Age20_25['Name'].count(), Age25_30['Name'].count(),Above30['Name'].count()])
mylabels = ['<=20','21-25','26-30','>30']
plt.title('Total players by Age')
plt.pie(x,labels = mylabels,autopct = '%.1f%%')
plt.show()
```

```python
#total under 20 players in each club
epl[epl['Age'] < 20]['Club'].value_counts().plot(kind = 'bar')
```

```python
#average age of players in each club
plt.figure(figsize = (12,6))
sns.boxplot(x= 'Club',y='Age',data = epl)
plt.xticks(rotation = 90)
```

```python
epl.groupby('Club')['Age'].mean().sort_values(ascending = False)
```

```python
#Total goals by each club
plt.figure(figsize = (12,6))
epl.groupby('Club')['Goals'].sum().sort_values(ascending = False).plot(kind = 'bar')
```

```python
#Top 10 players with most assists
top_10_assists = epl[['Name','Club','Assists','Matches']].sort_values(by= 'Assists',ascending = False).head(10)
top_10_assists
```

```python
#Top 10 goal scorers of the league 
top_10_scorers = epl[['Name','Club','Goals','Matches']].sort_values(by = 'Goals',ascending = False).head(10)
top_10_scorers
```

```python
top_10_goals_per_match = epl[['Name','Goals_per_match','Matches','Goals']].sort_values(by = 'Goals_per_match',ascending = False).head(10)
top_10_goals_per_match
```

```python
#Goals with assist and without assist
assists = epl['Assists'].sum()
goals = epl['Goals'].sum()
goal_wo_assists = goals - assists
plt.figure(figsize = (12,6))
data = [goals - assists, assists]
labels = ['Goals without assists', 'Goals with assists']
plt.pie(data,labels = labels,autopct = '%.0f%%')
```
